# Data Preprossing for the given files

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
#loading the data and data integration
df1 = pd.read_csv('Entertainer - Basic Info.csv')
df2 = pd.read_csv('Entertainer - Breakthrough Info.csv')
df3 = pd.read_csv('Entertainer - Last work Info.csv')
data = pd.merge(df1, df2, on='Entertainer')
data = pd.merge(data,df3, on='Entertainer')
data
```

```python
data = data.fillna(0)
data['Year of the First major achievement'] = data['Year of the First major achievement'].astype(int)
data['Year of Death'] = data['Year of Death'].astype(int)
data
```

```python
# Calculate additional features
data['Age at Breakthrough'] = data['Year of Breakthrough/#1 Hit/Award Nomination'] - data['Birth Year']
data['Career Span'] = data['Year of Last Major Work (arguable)'] - data['Year of Breakthrough/#1 Hit/Award Nomination']
data
```

# Visualization and Analysis

```python
#comparing between male and female actor
gender = data['Gender (traditional)'].value_counts()
gender.plot(kind ='pie',autopct = '%1.1f%%')
plt.title('Comparision between male and female')
plt.xlabel('')
plt.ylabel('')
plt.show()
```

```python
# Histogram for Year of Breakthrough
plt.hist(data['Year of Breakthrough/#1 Hit/Award Nomination'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Breakthrough Years')
plt.xlabel('Year of Breakthrough')
plt.ylabel('Number of Entertainers')
plt.show()
```

```python
# Sorting and selecting top 10 entertainers by career span
top_10_career_span = data.nlargest(10, 'Career Span')

plt.barh(top_10_career_span['Entertainer'], top_10_career_span['Career Span'])
plt.xlabel('Career Span (years)')
plt.title('Top 10 Entertainers by Career Span')
plt.gca().invert_yaxis()  
plt.show()
```

```python
# Histogram for Age at Breakthrough divided into bins
bins = range(10, 80, 10)
plt.hist(data['Age at Breakthrough'], bins=bins, color='salmon', edgecolor='black')
plt.title('Age at Breakthrough Distribution')
plt.xlabel('Age at Breakthrough')
plt.ylabel('Number of Entertainers')
plt.xticks(bins)
plt.show()
```

```python
# Box plot for Career Longevity by Gender
sns.boxplot(x='Gender (traditional)', y='Career Span', data=data, palette="pastel")
plt.title('Career Longevity by Gender')
plt.xlabel('Gender')
plt.ylabel('Career Span (years)')
plt.show()
```

```python
# Scatter plot for Age at Breakthrough vs Career Span
sns.scatterplot(x='Age at Breakthrough', y='Career Span', hue='Gender (traditional)', data=data, palette="deep")
plt.title('Age at Breakthrough vs Career Span')
plt.xlabel('Age at Breakthrough')
plt.ylabel('Career Span (years)')
plt.show()
```
